```python
i=0
encoder_list=[]
for index, row in data.iterrows():
    label={}
    commom= list(set(item).intersection(set(row)))
    uncommom=list(set(item)-set(row))
    for col in commom:
         label[col]=1
    for uncol in uncommom:
         label[uncol]=0
    encoder_list.append(label)
   
   
```

```python
encoder_list
```

```python
len(encoder_list)
```

```python
transaction=pd.DataFrame(encoder_list)
```

```python
transaction.head()
```

```python
frequent_items=apriori(transaction,
    min_support=0.5,
    use_colnames=True,
    max_len=True,
    verbose=0,
  )
```

```python
frequent_items
```
